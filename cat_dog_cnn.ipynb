```python
import numpy as np
import os
import matplotlib.pyplot as plt
import cv2
```

```python
cat_path = "PetImages\\Cat"
dog_path = "PetImages\\Dog"
```

```python
cat_list = os.listdir(cat_path)
cat_list
```

```python
dog_list = os.listdir(dog_path)
dog_list
```

```python
new_list = []
idx = []
def img_resize(path,liste,lengt_of_liste,mark):
    for i in range(lengt_of_liste):
        try:
            img = cv2.imread(path+"\\"+liste[i])
            img = cv2.resize(img,(128,128))        
            new_list.append(img)
            idx.append(mark)
        except:
            continue
```

```python
img_resize(cat_path,cat_list,2000,mark=0)
```

```python
for i in range(3):
    plt.imshow(new_list[i])
    plt.title(i)
    plt.show()
```

```python
img_resize(dog_path,dog_list,2000,1)
```

```python
for i in range(2000,2003):
    plt.imshow(new_list[i])
    plt.title(i)
    plt.show()
```

```python
X = new_list

Y = idx
```

```python
print(len(X),len(Y))
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train ,x_test ,y_train ,y_test = train_test_split(X,Y,train_size=0.8)
```

```python
x_train = np.array(x_train)
x_test = np.array(x_test)
y_train = np.array(y_train)
y_test = np.array(y_test)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
y_train = y_train.reshape(-1,1)
y_test = y_test.reshape(-1,1)
print(y_train.shape)
print(y_test.shape)
```

```python
x_train = x_train/255
x_train
```

```python
y_train
```

```python
x_test = x_test/255
x_test
```

```python
y_test
```

```python
import tensorflow as tf
```

```python
model = tf.keras.models.Sequential([
    tf.keras.layers.Conv2D(8,(5,5),activation="relu",input_shape=(128,128,3)),
    tf.keras.layers.MaxPooling2D((2,2)),
    tf.keras.layers.Conv2D(16,(3,3),activation="relu"),
    tf.keras.layers.MaxPooling2D((2,2)),
    
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(128,activation="relu"),
    tf.keras.layers.Dense(1,activation="sigmoid")
])
```

```python
model.summary()
```

```python
model.compile(optimizer="adam",loss="binary_crossentropy",metrics=["accuracy"])
```

```python
history = model.fit(x_train,y_train,epochs=5,validation_split=0.3)
```

```python
plt.plot(history.history["accuracy"],label="acc")
plt.plot(history.history["val_accuracy"],label="val_acc")
plt.legend()
plt.show()
```

```python
plt.plot(history.history["loss"],label="loss")
plt.plot(history.history["val_loss"],label="val_loss")
plt.legend()
plt.show()
```

```python
def model_prediction(img):
    img = cv2.imread(img)
    img = cv2.resize(img,(128,128))
    img = img/255
    plt.imshow(img)
    prediction = model.predict(img.reshape(1,128,128,3))
    print(prediction)
    print("Cat" if prediction < 0.5 else "Dog")
```
